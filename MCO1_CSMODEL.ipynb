## **4. Data Cleaning and Preprocessing**

!!! Add some intro here

!!! Add exploration of the dataset here (focused on analyzing the structure of the data source)

We execute '.info()' to examine column data types and non-null counts, and 'isnull.sum()' to identify missing values. 'describe()' provides summary statistics (mean, min, max, percentiles) for numeric column like views, subscribers, and rating. This initial audit reveals which columns need parsing or cleaning before analysis.

```python
# Strip all hidden spaces / line breaks from every text column
webtoons = webtoons.apply(lambda x: x.str.strip() if x.dtype == "object" else x)

# Array of possible null values seen in the dataset
null_strings = ['', ' ', 'null', 'Null', 'NULL', 'N/A', 'n/a']

# Apply the true Pandas null value to the suspected cells with empty values
webtoons = webtoons.replace(null_strings, np.nan)

# Check data types and missing values
webtoons.info()
print("\n\nMissing values per column:")
print(webtoons.isnull().sum())

# List of credit columns identified with missing values
credit_cols_to_drop = ['Created by','Written by', 'Art by', 'Adapted by', 'Original work by', 'Assisted by']

# Drop the columns from the DataFrame
webtoons = webtoons.drop(columns=credit_cols_to_drop)

# Convert released_date from object/string to datetime
webtoons['released_date'] = pd.to_datetime(webtoons['released_date'], errors='coerce')

# Summary statistics for numerical columns
webtoons.describe()
```

The output reveals the dataset structure: 734 rows and 19 columns.

<br>

**Key observations:**  
- Numeric columns (`views`, `subscribers`, `grade`, `episodes`, `likes`) are properly typed.
- The credit columns (`written_by`, `art_by`, `adapted_by`, etc.) contain many missing values.
- The `grade` column contains real rating values (not all zeros), enabling us to analyze reader ratings.
- The `released_date` column allows us to calculate time on platform.
